```python
import os
import zipfile
import pandas as pd

# Function to extract and read data from a zip file
def read_zip_data(zip_file_path):
    num_data = []
    sub_data = []

    with zipfile.ZipFile(zip_file_path, 'r') as zip_file:
        with zip_file.open('num.txt') as num_file:
            num_data = pd.read_csv(num_file, sep='\t')
        with zip_file.open('sub.txt') as sub_file:
            sub_data = pd.read_csv(sub_file, sep='\t')

    return num_data, sub_data

# Input path containing multiple zip files
input_path = "gdrive/My Drive/Colab Notebooks/SEC_Data/Fraud"

num_dataframes = []
sub_dataframes = []

Fraud_label_df = pd.read_csv("gdrive/My Drive/Colab Notebooks/SEC_Data/Fraud_Labels.csv")

# Loop through zip files in the input directory
for root, dirs, files in os.walk(input_path):
    for file in files:
        if file.endswith(".zip"):
            zip_file_path = os.path.join(root, file)
            num_df, sub_df = read_zip_data(zip_file_path)
            num_dataframes.append(num_df)
            sub_dataframes.append(sub_df)

# Concatenate dataframes from all zip files
if num_dataframes:
    num_dataframe = pd.concat(num_dataframes, ignore_index=True)
    num_dataframe = num_dataframe[num_dataframe.adsh.isin(Fraud_label_df.adsh)]
    sub_dataframe = pd.concat(sub_dataframes, ignore_index=True)
    sub_dataframe = sub_dataframe[sub_dataframe.adsh.isin(Fraud_label_df.adsh)]


    # You can now work with the num_dataframe and sub_dataframe as needed
    print("Dataframes for num.txt and sub.txt are created.")
    print("num_dataframe:")
    print(num_dataframe.head())
    print("sub_dataframe:")
    print(sub_dataframe.head())
else:
    print("No zip files found in the specified directory.")

```
